```python
import folium
import pandas as pd

ma_geo = 'ma.json'
uber_waittimes = 'csv/uber.csv'
uber_data = pd.read_csv(uber_waittimes)
```

```python
#lambda does the same thing as the following for loop
#counties = []
#for idx, row in uber_data.iterrows():
#    counties.append('05000US' + str(int(row.County)))

uber_data['County'] = uber_data.apply(lambda row: '05000US' + str(int(row.County)), axis=1)
#uber_data['AverageWaitTime'] = uber_data.apply(lambda row: row['AverageWaitTime']/60.0, axis=1) #messes up the color
```

```python
#Let Folium determine the scale
mass_map = folium.Map(location=[42.369444, -71.948611], zoom_start=7) #this is hard coded - still looking for a dataset
mass_map.geo_json(geo_path=ma_geo, data=uber_data,
             columns=['County', 'AverageWaitTime'],
             key_on='feature.properties.geoid',
             data_out='data.json',
             #threshold_scale=[minimum, first_quarter, mean, second_quarter],
             fill_color='YlGn', fill_opacity=0.7, line_opacity=0.2,
             legend_name='Average Wait Time for an Uber (sec)')
mass_map.create_map(path='mass_uber.html')
```

```python
#http://nbviewer.ipython.org/url/ocefpaf.github.com/python4oceanographers/downloads/notebooks/2015-02-02-cartopy_folium_shapefile.ipynb
#This function is created by Filipe Fernandes and found at the above link, used under CC:Attribution-ShareAlike. No modifications have been made.
from IPython.display import IFrame, HTML

def inline_map(m, width=650, height=500):
    """Takes a folium instance and embed HTML."""
    m._build_map()
    srcdoc = m.HTML.replace('"', '&quot;')
    embed = HTML('<iframe srcdoc="{}" '
                 'style="width: {}px; height: {}px; '
                 'border: none"></iframe>'.format(srcdoc, width, height))
    return embed
```
