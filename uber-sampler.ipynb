```python
import pandas as pd
#initilize uber API
from uberpy import Uber
with open('keys/uber.txt','r') as f:
        client_id = f.readline().strip()
        server_token = f.readline().strip()
        secret= f.readline().strip()
uber = Uber(client_id, server_token, secret)
#1000 requests per hour
```

```python
df = pd.DataFrame.from_csv("csv/samples.csv")
df = df[df['states'] == 'MA']
```

```python
df
```

```python
def uberize(row):
    #take in a pair of lat and lng as two columns
    #return average wait time for any uber pick up
    total = []
    for idx,sample in enumerate(row):
        #clean = (sample.replace(' ','').replace('-',',-').replace('[','').replace(']','')).split(',') #need to fix
        try:
            clean = eval(sample)
        except NameError: #this is the 'states' column
            pass
        times = uber.get_time_estimate(clean[0], clean[1])
        for x in (times['times']):
            total.append(x['estimate'])
    try:
        return sum(total)/float(len(total))
    except: #this should only catch 'Too Many Requests'
        return -1
```

```python
var = df.apply(lambda d: uberize(d), axis=1)
```

```python
print var
```

```python
v = pd.DataFrame(var)
```

```python
v['County'] = v.index
v
```

```python
v.to_csv("csv/uber.csv", header=['AverageWaitTime', 'County'], index=False)
```
